1) What is the difference between descriptive statistics and inferential statistics? Explain with examples.

  -  Descriptive statistics summarize known data (mean, median, charts), while inferential statistics use sample data to predict or generalize about a larger population (hypothesis tests, confidence intervals), focusing on "what is" versus "what could be". Descriptive stats describe your data; inferential stats infer about a bigger group.


2)What is sampling in statistics? Explain the differences between random and stratified sampling.

- Sampling in statistics selects a subset (sample) from a larger group (population) to make inferences about the whole, with simple random sampling giving everyone an equal chance, while stratified sampling first divides the population into key subgroups (strata, e.g., age, gender) and then samples randomly from each stratum, ensuring representation of all segments for more precise results, especially when subgroups differ significantly.

3) Define mean, median, and mode. Explain why these measures of central tendency are important.

- Mean, median, and mode are statistical measures describing a dataset's center: the mean is the average (sum/count), the median is the middle value in an ordered list, and the mode is the most frequent value; they are vital for summarizing data, identifying typical values, understanding distribution (especially with outliers where median is best), and making informed decisions in fields like business, finance, and research.

4) Explain skewness and kurtosis. What does a positive skew imply data?

- Skewness measures a distribution's asymmetry, indicating if data leans left or right, while kurtosis measures its tailedness/peakedness, showing how heavy or light the tails are compared to a normal curve, revealing outlier presence. A positive skew means the tail stretches to the right, implying most data points are lower, with a few unusually high values pulling the mean up past the median.

```python

# 5) Implement a Python program to compute the mean, median, and mode of a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
from collections import Counter
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean_value = sum(numbers) / len(numbers)

# Median
median_value = statistics.median(numbers)

# Mode (handles one or more modes)
frequency = Counter(numbers)
max_count = max(frequency.values())
mode_value = [num for num, count in frequency.items() if count == max_count]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
#7) Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:

#data[12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert to NumPy array
arr = np.array(data)

# Calculate quartiles and IQR
Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
IQR = Q3 - Q1

# Calculate outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = arr[(arr < lower_bound) | (arr > upper_bound)]

# Draw boxplot
plt.boxplot(arr, vert=False)
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()

# Print results
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {list(outliers)}")
```

```python
""" 8) You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.

Explain how you would use covariance and correlation to explore this relationship.

Write Python code to compute the correlation between the two lists:

advertising_spend = [200, 250, 300, 400, 500]

daily_sales [2200, 2450, 2750, 3200, 4000]"""

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute the Pearson correlation coefficient
correlation_matrix = np.corrcoef(advertising_spend, daily_sales)
correlation = correlation_matrix[0, 1]

print("Correlation between advertising spend and daily sales:", correlation)
```
